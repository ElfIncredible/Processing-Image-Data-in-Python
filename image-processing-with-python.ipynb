**Resizing the image: Pillow library**

**Convert image to grayscale: OpenCV**

```python
img = cv2.imread('/kaggle/working/bmw-3-series.jpg')
```

```python
type(img)
```

```python
img.shape
```

```python
# convert the image from BGR to RGB

img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
# convert image from the RGB color space to a grayscale image

img_grayscale = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2GRAY)
```

```python
type(img_grayscale)
```

```python
img_grayscale.shape
```

```python
# Display the grayscale image

plt.imshow(img_grayscale, cmap='gray')  
plt.show()
```
